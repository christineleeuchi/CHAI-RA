```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Persuasive Argument Pairs Valence Scores Per Quarter
persuasive_argument_valence_scores_df = pd.read_csv("pairs_persuasiveness_analysis/pairs_linguistics_scores/valence/persuasive_quarter_scores.csv")
persuasive_argument_valence_quarter_score_averages_df = persuasive_argument_valence_scores_df.mean()

# Non-Persuasive Argument Pairs Valence Scores Per Quarter
non_persuasive_argument_valence_scores_df = pd.read_csv("pairs_persuasiveness_analysis/pairs_linguistics_scores/valence/non_persuasive_quarter_scores.csv")
non_persuasive_argument_valence_quarter_score_averages_df = non_persuasive_argument_valence_scores_df.mean()

plt.plot(persuasive_argument_valence_quarter_score_averages_df, label="effective")
plt.plot(non_persuasive_argument_valence_quarter_score_averages_df, label="non-effective")
plt.legend(loc="upper left")
plt.title("Average Valence in Root Replies by Quarter")
plt.show()
```
